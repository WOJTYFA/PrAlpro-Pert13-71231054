```python
#latihan 12.1
def main():
    n = int(input("Masukan jumlah kategorinya: "))

    categories = {}

    for i in range(n):
        kategori = input(f"Masukan nama kategori ke-{i+1}: ")
        apps = []
        print(f"Masukan 5 aplikasi untuk kategori '{kategori}: ")
        for j in range(5):
            app = input(f"Aplikasi {j+1}: ")
            apps.append(app)
        categories[kategori] = set(apps)

    print("\nDaftar nama aplikasi di setiap kategori: ")
    for kategori, apps in categories.items():
        print(f"{kategori}: {','.join(apps)}")

    semua_categories_apps = set.intersection(*categories.values())
    print("\nAplikasi yang muncul di semua kategori: ")
    if semua_categories_apps:
        print(", ".join(semua_categories_apps))
    else:
        print("Tidak ada aplikasi yang muncul di semua kategori")

    itung_apk = {}
    for apps in categories.values():
        for apk in apps:
            if apk in itung_apk:
                itung_apk[apk] += 1
            else:
                itung_apk[apk] = 1
    
    satu_kategoti_apps = {app for app, count in itung_apk.items() if count == 1}
    print("\nAplikasi yang hanya muncul di satu kategori: ")
    if satu_kategoti_apps:
        print(",".join(satu_kategoti_apps))
    else:
        print("Tidak ada aplikasi yang hanya muncul di satu kategori")

    if n > 2:
        dua_kategori_apps = {app for app, count in itung_apk.items() if count == 2}
        print("\nAplikasi yang muncul di dua kategori: ")
        if dua_kategori_apps:
            print(",".join(dua_kategori_apps))
        else:
            print("Tidak ada aplikasi yang muncul tepat dua kategori.")
    else:
        print("\nJumlah kategori > dari 3, tidak ada aplikasi yang muncul tepat di dua kategori")

```

```python
#latihan 12.2
def list_ke_set(lst):
    print("List sebelum dikonversi", lst)
    konversi_set = set(lst)
    print("Set setelah dikonversi", konversi_set)
    return konversi_set
def set_ke_list(st):
    print("Set sebelum dikonversi", st)
    konversi_list = list(st)
    print("List setelah dikonversi", konversi_list)
    return konversi_list
def tuple_ke_set(tpl):
    print("Tuple sebelum dikonversi", tpl)
    konversi_set = set(tpl)
    print("Set setelah dikonversi", konversi_set)
    return konversi_set
def set_ke_tuple(st):
    print("Set sebelum konversi", st)
    konversi_tuple = tuple(st)
    print("Tuple setelah dikonversi", konversi_tuple)
    return konversi_tuple
```
